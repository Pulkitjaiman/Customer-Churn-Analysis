```python
plt.figure(figsize = (3, 4))
gb = df.groupby("Churn").agg({'Churn':"count"})
plt.pie(gb['Churn'], labels = gb.index, autopct = "%1.2f%%")
plt.title("Percentage of Churned Customers", fontsize = 10)
plt.show()
```

```python
# Calculate counts
grouped = df.groupby(['SeniorCitizen', 'Churn']).size().unstack(fill_value=0)

# Convert to percentages
percentages = grouped.div(grouped.sum(axis=1), axis=0) * 100

# Plot
labels = percentages.index.astype(str)
yes = percentages['Yes']
no = percentages['No']

plt.figure(figsize=(4, 4))
bar1 = plt.bar(labels, no, label='No', color='skyblue')
bar2 = plt.bar(labels, yes, bottom=no, label='Yes', color='grey')

# Add percentage labels
for i in range(len(labels)):
    plt.text(i, no.iloc[i] / 2, f"{no.iloc[i]:.1f}%", ha='center', va='center', fontsize=9)
    plt.text(i, no.iloc[i] + yes.iloc[i] / 2, f"{yes.iloc[i]:.1f}%", ha='center', va='center', fontsize=9)

plt.title('Churn by SeniorCitizen (in %)')
plt.xlabel('SeniorCitizen')
plt.ylabel('Percentage')
plt.legend(title='Churn')
plt.tight_layout()
plt.show()
```

```python
cols = [
    'PhoneService', 'MultipleLines', 'InternetService',
    'OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
    'TechSupport', 'StreamingTV', 'StreamingMovies'
]

n_cols = 3
n_rows = -(-len(cols) // n_cols)

plt.figure(figsize=(5 * n_cols, 4 * n_rows))

for i, col in enumerate(cols, 1):
    plt.subplot(n_rows, n_cols, i)
    sns.countplot(data=df, x=col,  palette='pastel', legend=False, hue = df["Churn"])
    plt.title(col)
    plt.xticks(rotation=45)
    plt.tight_layout()

plt.suptitle("Countplots for Service Features", fontsize=16, y=1.02)
plt.tight_layout()
plt.show()
```

Most customers have active PhoneService, with relatively few having no phone service. Features like OnlineSecurity, OnlineBackup, TechSupport, and DeviceProtection show a noticeable drop-off in usage among customers who churned. Services requiring internet (like StreamingTV, StreamingMovies) show fewer churners among those who use them, while "No internet service" often corresponds to low churn and lower engagement overall.
